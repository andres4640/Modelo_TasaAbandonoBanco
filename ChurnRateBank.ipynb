# Clientes Tarjetas de Credito

El adminsitrador del banco se encuentra preocupado debido a que varios clientes se encuentran abandonando los servicios de tarjeta de credito. Debemos predecir que clientes van a dejar los servicios para proveerles mejores ofertas y/o servicios y de esta forma cambiar su decisión.

Muchas empresas gastan dinero en sus esfuerzos de adquirir nuevos clientes. Si estos clientes deciden no quedarse por mucho tiempo, esto quiere decir que el gasto para adquirir nuevos clientes tiene un ROI (Return of Investment) negativo.

## Objetivo
- Crear un Modelo de Clasficación que permita predecir que cliente abandonara los servicios de tarjeta de credito.

## Palabras Clave:
- Desgaste de clientes: Perdida de clientes de un negocio. 
- Ratio de Desgaste de clientes: Dividir el numero de clientes que tenia al comienzo de un periodo entre el numero que tendra al final del periodo
- Months on book: Respecto a un Préstamo Adquirido, el número de meses naturales completos que han transcurrido desde el mes natural en el que se produjo dicho Préstamo, medidos al final de cada mes natural.

# Modelo de Clasificacion

## Pre procesamiento de datos

Antes de crear el modelo de predicción, debemos realizar un procesamiento previo de los datos para que estos puedan ser leidos por el modelo. Seguiremos los siguientes pasos:
1. Crear dos litas con el nombre de las variables numericas y categoricas.
2. Aplicar la tecnica One Hot Encoding en las variables categoricas para que puedan ser interpretadas por el modelo, ya que no puede interpretar texto.
3. Cambiaremos los valores de la variable objetivo, "Attrition_Flag", de texto a numeros. Los clientes actuales recibiran el valor de 0 y 1 en caso sea un cliente perdido.
4. Aplicaremos la tecnica de escalación a las variables numericas para que todas se encuentren en una misma escala y no permitir que alguna variable con mayores valores numericos dominen a las demas variables. 

```python
dataset.shape
```

```python
df = dataset.copy()
```

```python
df["Target"] = df["Attrition_Flag"].apply(lambda x: 0 if x == "Existing Customer" else 1)
df = df.drop(columns=["Attrition_Flag"])
df = df.astype({"Target": "int64"})
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

### Dividimos dataset en variables categoricas y numericas

```python
columnas_cat = ["Gender","Education_Level", "Marital_Status", "Income_Category", "Card_Category"]

columnas_cont = [ ele for ele in list(x.columns) ] 
for a in columnas_cat: 
  if a in list(x.columns): 
    columnas_cont.remove(a) 
```

### Transformamos las variables categoricas

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

x_cat = x[columnas_cat].values
x_con = x[columnas_cont].values

ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(sparse=False), [0,1,2,3,4])], remainder="passthrough")
x_cat = np.array(ct.fit_transform(x_cat))


x_final = np.concatenate([x_cat,x_con], axis=1)

le = LabelEncoder()
y = le.fit_transform(y)

x_train, x_test, y_train, y_test = train_test_split(x_final, y, test_size = 0.2)
```

### Escalación de variables numericas

```python
sc = StandardScaler()
x_train[:,22:] = sc.fit_transform(x_train[:,22:])

x_test[:,22:] = sc.transform(x_test[:,22:])
```

## Balancear Dataset

Aplicaremos la tecnica SMOTETomek el cual nos permitira crear ejemplos sinteticos para la categoria con menor cantidad de valores (Clientes perdidos) y aplicaremos una tecnica de undersampling (Tomek Link) para reducirla cantidad de ejemplos de la categoria mayoritaria (Clientes Existentes). Podemos ver que la diferencia de ejemplos de la clase 0, clientes actuales, y la clase 1, clientes perdidos, es grande. Especificaremos que el numero de ejemplos de clientes perdidos sea proporcional al 80% del numero de ejemplos de los clientes actuales.

```python
from collections import Counter
print("Clases: {}".format((Counter(y_train))))
```

```python
from imblearn.combine import SMOTETomek

smt = SMOTETomek(sampling_strategy=0.8)
x_train, y_train = smt.fit_resample(x_train, y_train)
```

```python
print("Clases: {}".format((Counter(y_train))))
```

Podemos visualizar que actualmente en el dataset de entrenamiento tenemos 6822 ejemplos que son de clientes actuales y 1279 de clientes perdidos. Mediante la aplicación de la tecnica SMOTETomek, pudimos balancear el dataset para que la diferencia entre el numero de ejemplos de cada clase no sea tan grande.

# Entrenar Modelos

Nuestros datos ya se encuentra procesados y listo para usarlos en el entrenamiento de nuestros modelos de predicción. Entrenaremos diferentes modelos y evaluaremos cual es el más adecuado para nuestro caso de estudio. Usaremos metricas como "accuracy", "confusion matrix" y la curva AUROC para evaluar el performance de cada modelo.

## Funcion Evaluar Modelos

Crearemos una función que permita calculare la metrica "accuracy" y la "confusion matrix" para evaluar el performance de cada modelo.

```python
from sklearn.metrics import confusion_matrix, accuracy_score

def evaluar_modelo(y_pred):
    matrix = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in
                    matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in
                         matrix.flatten()/np.sum(matrix)]
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    ax = sns.heatmap(matrix, annot=labels, fmt='', cmap='Blues')
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    plt.show()
    
    print("\nAccuracy: ", accuracy)
```

```python
train = 1
```

## Regresion Logistica

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
resultados = {}
```

```python
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
```

```python
from sklearn.linear_model import LogisticRegression

resultado = cross_val_score(LogisticRegression(), x_train, y_train, scoring = "accuracy", cv=10)
resultados["Logistic_Reg"] = {"Media": np.mean(resultado), 
                                   "Min": np.min(resultado),
                                   "Max": np.max(resultado),
                                  "Std": np.std(resultado)}
print(resultados["Logistic_Reg"])
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)

lr_pred = lr.predict(x_test)

evaluar_modelo(lr_pred)
```

## Support Vector Machine

```python
from sklearn.svm import SVC

if train != 1:
    parametros=[{"C":[0.25,0.5,0.75,1], "kernel":["linear"]},
               {"C":[0.25,0.5,0.75,1], "kernel":["rbf"], "gamma":np.arange(0.1,1,0.1)}]

    grid_search=GridSearchCV(estimator=SVC(),
                            param_grid = parametros,
                            scoring = "accuracy",
                            cv = 10,
                            n_jobs = -1
                            )
    grid_search.fit(x_train, y_train)
    
    best_accuracy = grid_search.best_score_
    best_parameters = grid_search.best_params_

    print("Mejor accuracy: {:.2f} %".format(best_accuracy*100))
    print("Mejores parametros: ", best_parameters)
```

```python
if train != 1:
    svc = SVC(C = best_parameters["C"], gamma = best_parameters["gamma"], kernel= best_parameters["kernel"])
else:
    svc = SVC(C = 1, gamma = 0.1, kernel= 'rbf', probability=True)


resultado = cross_val_score(svc, x_train, y_train, scoring = "accuracy", cv=10)
resultados["SVC"] = {"Media": np.mean(resultado), 
                                   "Min": np.min(resultado),
                                   "Max": np.max(resultado),
                                  "Std": np.std(resultado)}
print(resultados["SVC"])
```

```python
svc.fit(x_train, y_train)
svc_pred = svc.predict(x_test)
evaluar_modelo(svc_pred)
```

## Clasificador KNeighbors

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors=5, metric = "minkowski", p=2)

resultado = cross_val_score(kn, x_train, y_train, scoring = "accuracy", cv=10)
resultados["KNN"] = {"Media": np.mean(resultado), 
                                   "Min": np.min(resultado),
                                   "Max": np.max(resultado),
                                  "Std": np.std(resultado)}
print(resultados["KNN"])
```

```python
kn.fit(x_train, y_train)
knn_pred = kn.predict(x_test)
evaluar_modelo(knn_pred)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

if train != 1:
    parametros=[{"n_estimators": np.arange(100, 1000, 50),
                "criterion":["gini", "entropy"],
                 "bootstrap" : [True, False],
                 "max_features": ["auto", "sqrt", "log2"]
                }]

    grid_search=GridSearchCV(estimator=RandomForestClassifier(),
                            param_grid = parametros,
                            scoring = "accuracy",
                            cv = 10,
                            n_jobs = -1
                            )
    grid_search.fit(x_train, y_train)
    
    best_accuracy = grid_search.best_score_
    best_parameters = grid_search.best_params_

    print("Mejor accuracy: {:.2f} %".format(best_accuracy*100))
    print("Mejores parametros: ", best_parameters)
```

```python
if train != 1:
    rf = RandomForestClassifier(n_estimators=best_parameters["n_estimators"], 
                               bootstrap=best_parameters["bootstrap"],
                               criterion=best_parameters["criterion"],
                               max_features=best_parameters["max_features"])
else:
    
    rf = RandomForestClassifier(n_estimators=850, 
                           bootstrap=False,
                           criterion="gini",
                           max_features="auto")
resultado = cross_val_score(rf, x_train, y_train, scoring = "accuracy", cv=10)
resultados["RandomForest"] = {"Media": np.mean(resultado), 
                              "Min": np.min(resultado),
                              "Max": np.max(resultado),
                              "Std": np.std(resultado)}
print(resultados["RandomForest"])
```

```python
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
evaluar_modelo(rf_pred)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(x_train, y_train)

resultado = cross_val_score(gnb, x_train, y_train, scoring = "accuracy", cv=10)
resultados["NaiveBayes"] = {"Media": np.mean(resultado), 
                               "Min": np.min(resultado),
                               "Max": np.max(resultado),
                               "Std": np.std(resultado)}
print(resultados["NaiveBayes"])

gnb_pred = gnb.predict(x_test)
evaluar_modelo(gnb_pred)
```

## Reporte de Clasificación

No nos podemos basar solo en la metrica accuracy, nostros estamos interesados en saber que tan preciso fue el modelo al clasificar los ejemplos que si eran clientes perdidos. Para esto, usaremos el reporte de clasificación para conocer las demas metricas de los modelos entrenados. Entre ellas la de nuestro interes es el "recall" y "f1-support".

```python
from sklearn.metrics import classification_report


print('Regresión Logistica:')
print(classification_report(y_test, lr_pred))

print('KNears Neighbors:')
print(classification_report(y_test, knn_pred))

print('Support Vector Machine:')
print(classification_report(y_test, svc_pred))

print('Random Forest:')
print(classification_report(y_test, rf_pred))

print('Naive Bayes:')
print(classification_report(y_test, gnb_pred))
```

# ROC Curve

La ultima metrica que usaremos el area debajo de la curva ROC. Esta metrica combina el  TPR (True Positive Rate) y FPR (False Positive Rate) para ver el performance de cada modelo. Usaremos la probabilidad de predicción de cada uno de los modelos al predecir los datos de prueba. Tambien se podra visualiar como evolucionan el TPR y FPR en diferentes umbrales de clasificación.

Queremos maximizar el TPR y minimizar el FPR.

```python
r_probs = [0 for _ in range(len(y_test))]
rf_probs = rf.predict_proba(x_test)
lr_probs = lr.predict_proba(x_test)
gnb_probs = gnb.predict_proba(x_test)
kn_probs = kn.predict_proba(x_test)
svc_probs = svc.predict_proba(x_test)

rf_probs = rf_probs[:, 1]
lr_probs = lr_probs[:, 1]
gnb_probs = gnb_probs[:, 1]
kn_probs = kn_probs[:, 1]
svc_probs = svc_probs[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

r_auc = roc_auc_score(y_test, r_probs)
rf_auc = roc_auc_score(y_test, rf_probs)
lr_auc = roc_auc_score(y_test, lr_probs)
gnb_auc = roc_auc_score(y_test, gnb_probs)
kn_auc = roc_auc_score(y_test, kn_probs)
svc_auc = roc_auc_score(y_test, svc_probs)
```

## AUROC Scores

```python
print('Random (chance) Prediction: AUROC = %.3f' % (r_auc))
print('Random Forest: AUROC = %.3f' % (rf_auc))
print('Linear Regression: AUROC = %.3f' % (lr_auc))
print('Naive Bayes: AUROC = %.3f' % (gnb_auc))
print('SVM: AUROC = %.3f' % (svc_auc))
print('KNN: AUROC = %.3f' % (kn_auc))
```

```python
r_fpr, r_tpr, _ = roc_curve(y_test, r_probs)
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
gnb_fpr, gnb_tpr, _ = roc_curve(y_test, gnb_probs)
kn_fpr, kn_tpr, _ = roc_curve(y_test, kn_probs)
svc_fpr, svc_tpr, _ = roc_curve(y_test, svc_probs)

plt.figure(figsize=(12,7))
plt.plot(r_fpr, r_tpr, linestyle='--', label='Random prediction (AUROC = %0.3f)' % r_auc)
plt.plot(rf_fpr, rf_tpr, marker='.', label='Random Forest (AUROC = %0.3f)' % rf_auc)
plt.plot(gnb_fpr, gnb_tpr, marker='.', label='Naive Bayes (AUROC = %0.3f)' % gnb_auc)
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic Regression (AUROC = %0.3f)' % lr_auc)
plt.plot(kn_fpr, kn_tpr, marker='.', label='KNN (AUROC = %0.3f)' % kn_auc)
plt.plot(svc_fpr, svc_tpr, marker='.', label='SVM (AUROC = %0.3f)' % svc_auc)

plt.title('ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
